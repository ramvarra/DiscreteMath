```python
def offset(c, key):
    if not c.isalpha():
        return c
    shift = (ord(c) - ord(key)  + 26) % 26
    return chr((ord('A') + shift))

def decode(text, key):
    return ''.join(offset(c, key) for c in text)
```

```python
for i in range(1, 26):
    key = chr(ord('A') + i)
    print(key, ': ', decode(ct, key))
```
